# Imported necessary libraies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Read the data file

```python
df = pd.read_csv("DoctorVisits - DA.csv")
df.head(15)
```

# Displayed complete information about the columns of the dataset

```python
df.info()
```

# Find total number of people based on illness

```python
df["illness"].value_counts()
```

```python
df["gender"].value_counts()
```

# Visualized and analysed the max,min and medium income

```python
y = list(df.income)
plt.boxplot(y)
plt.show()
```

# No.of days of reduced activity of male and female separatly due to illness

```python
df.groupby(["gender","reduced"]).mean()
```

# Find is there any missing values using heat map

```python
sns.heatmap(df.isnull(),cbar=False,cmap='viridis')
```

# Correlation between different variables

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='Blues')
```

# How income of patients affects no.of visits.

```python
plt.figure(figsize=(10,10))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('vistis')
```

# No.of males and females affected by illness

```python
sns.histplot(df.gender,bins=2)
```

# Visualize the percentage of people getting govt health insurance due to low income, old age and having private insurance

```python
# % of people getting govt insurance due to low income
label = ['yes','no']
y = df[df['freepoor']=='yes']
n = df[df['freepoor']=='no']
x = [y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt insurance due to low income")
plt.show()
```

```python
# % of people having private Insurance
y = df[df['private']=='yes']
n = df[df['private']=='no']
x = [y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people having private Insurance")
plt.show()
```

```python
# % of people getting govt insurance due to old age, disability
y = df[df['freerepat']=='yes']
n = df[df['freerepat']=='no']
x = [y.shape[0],n.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title("% of people getting govt insurance due to old age, disability")
plt.show()
```

# Reduced days of activity due to illness based on gender

```python
db = df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color=['cornflowerblue','lightseagreen'])
plt.title('Bar chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
plt.show()
```
